### Compute Topic Models
We will load each of the news article CSV files again into dataframes and then process the `content` column as our documents and the `id` column as our document IDs.

After each model is processed, we will save the generated model to the `model_output` directory. 

```python
from top2vec import Top2Vec
import pandas as pd 
```

```python
# define a function that will take in a list of articles and IDs and compute a topic model 
def compute_topic_model(documents=None, ids=None): 
  assert documents is not None, "Must supply valid list of documents"
  assert ids is not None, "Must supply a valid list of document IDs"
  assert len(documents) == len(ids), "Documents and Document IDs must be the same shape"
  model = Top2Vec(documents=documents, document_ids=ids, embedding_model='all-MiniLM-L6-v2',
                  use_embedding_model_tokenizer=True, keep_documents=False)
  return model 
```

```python
# Reload the data frames and compute a topic model for each 
news1_df = pd.read_csv('articles1.csv')
news2_df = pd.read_csv('articles2.csv')
news3_df = pd.read_csv('articles3.csv')
news1_df.head()
```

```python
# Grab the list of ids and content to use to compute the topic model 
docs = list(news1_df['content'])
ids = list(news1_df['id'])
len(docs), len(ids)
```

```python
news1_model = compute_topic_model(docs, ids)
```

```python
news1_model_file = '../model_output/articles1.model'
news1_model.save(news1_model_file)
```

```python
docs = list(news2_df['content'])
ids = list(news2_df['id'])
len(docs), len(ids)
```

```python
news2_model = compute_topic_model(docs, ids)
```

```python
news2_model_file = '../model_output/articles2.model'
news2_model.save(news2_model_file)
```

```python
docs = list(news3_df['content'])
ids = list(news3_df['id'])
len(docs), len(ids)
```

```python
news3_model = compute_topic_model(docs, ids)
```

```python
news3_model_file = '../model_output/articles3.model'
news3_model.save(news3_model_file)
```
